# approximate radius of earth in km

```python
from math import sin, cos, sqrt, atan2, radians
```

```python
def dist(host_lat, host_long, t_lat, t_long):


    R = 6373.0

    lat1 = radians(host_lat)
    lon1 = radians(host_long)
    lat2 = radians(t_lat)
    lon2 = radians(t_long)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    
    return distance 

def dist_series(host_lat, host_long, t_lat, t_long):
    
    distances = []
    
    for i in range(len(host_lat)):
        distance = dist(host_lat[i], host_long[i], t_lat[i], t_long[i])
        distances.append(distance) 
        
    return pd.Series(distances)
```

```python
dfr['t1_dist_home'] = dist_series(dfr['host_lat'] ,dfr['host_long'],dfr['team1_lat'],dfr['team1_long'])
dfr['t2_dist_home'] = dist_series(dfr['host_lat'] ,dfr['host_long'],dfr['team2_lat'],dfr['team2_long'])
```

```python
def z_norm(x):
    
    res = (x - x.mean()) / x.std()
    
    return res
```

```python
dfr['t1_dist_home'] = z_norm(dfr['t1_dist_home'])
dfr['t2_dist_home'] = z_norm(dfr['t2_dist_home'])
```
