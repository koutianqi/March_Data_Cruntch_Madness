# Import Raw Data

```python
import pandas as pd, numpy as np
```

```python
dfr = pd.read_csv('NCAA_Tourney_2002-2017.csv')
target = dfr['result']
df = dfr.drop('result', axis = 1)
```

```python
print(list(df.columns))
```

# Dummy Up

### team1_teamname

```python
team1_id = pd.get_dummies(df['team1_id'])
team1_id.head()
```

### team1_teamname

```python
team2_id = pd.get_dummies(df['team2_id'])
team2_id.head()
```

### team1_coach_id

```python
team1_coach_id = pd.get_dummies(df['team1_coach_id'])
team1_coach_id.head()
```

### team1_coach_id

```python
team2_coach_id = pd.get_dummies(df['team2_coach_id'])
team2_coach_id.head()
```

### Put back

```python
df1 = team1_id.merge(team2_id, left_index=True, right_index = True)
```

```python
df2 = df1.merge(team1_coach_id, left_index=True, right_index = True)
```

```python
df3 = df2.merge(team2_coach_id,left_index=True, right_index = True )
```
